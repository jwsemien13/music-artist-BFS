```python
#import libraries
import networkx as netX
import matplotlib.pyplot as vis
import time
import memory_profiler as mem_profile

class MusicBFS_Navigator:   
    #Build network using adjacency matrix            
    def build_network(self):                   
        self.network = { 'Drake1': ['LilBaby', 'NickiMinaj', 'Offset'],
                       'LilBaby': ['Drake2', 'RodWave', 'TravisScott', 'Gunna'],
                    'NickiMinaj': ['Drake3', 'BabyfaceRay'],
                        'Offset': ['Veeze', 'Future'],
                        'Drake2': [],
                       'RodWave': [],
                   'TravisScott' : [], 
                         'Gunna': [],                                                                                                
                        'Drake3': [],
                   'BabyfaceRay': [],
                        'Veeze' : [],
                        'Future': []}
                      
    def visualize_Network(self): 
        net = netX.Graph()
        net.add_edge('Drake', 'LilBaby')          
        net.add_edge('Drake', 'NickiMinaj') 
        net.add_edge('Drake', 'Offset') 
        
        net.add_edge('LilBaby', 'Drake2') 
        net.add_edge('LilBaby', 'RodWave') 
        net.add_edge('LilBaby', 'TravisScott')      
        net.add_edge('LilBaby', 'Gunna')        
        
        net.add_edge('NickiMinaj', 'Drake3')      
        net.add_edge('NickiMinaj', 'BabyfaceRay')   
        
        net.add_edge('Offset', 'Veeze')      
        net.add_edge('Offset', 'Future')   
        
        netX.draw(net, with_labels=True)  
        vis.show()
     
    def BF_Search(self, vertex, destination): 
                  
        explored = []           # Initalize 
          
        FIFO_queue = []          
        FIFO_queue.append(vertex) 
         
        while FIFO_queue:               
            vertex = FIFO_queue.pop(0)    #shallowest vertex
            
            if vertex not in explored:
               explored.append(vertex)
               if vertex == destination:
                  return explored
              
               frontiers = self.network[vertex]                
               for frontier in frontiers:
                    FIFO_queue.append(frontier)                   
                     
        return "Path not found"




# -----------Driving the above class ----------------------------------- 

print('N A V I G A T O R: Breadth-first Search')
#Start Time and memory complexity           

t1 = time.process_time()                    #Time
m1 = mem_profile.memory_usage()             #Memory

bfs = MusicBFS_Navigator() 
#Build Network
bfs.build_network() 

#Visualize network
bfs.visualize_Network()

# Searching from initial vertex n and display path
initial_state = 'Drake1'
print('Root State = ', initial_state)
destination = str(input("Enter Destination State: "))

explored = bfs.BF_Search(initial_state, destination) 
print('\n Path found from: ', initial_state, ' to: ', destination)
print(explored)

#End Time and memory 

t2 = time.process_time()
print('\n Time complexity = ', abs(t2-t1), ' seconds')

m2 = mem_profile.memory_usage()
print('\n Memory complexity = ', m2[0] - m1[0], ' Mb')      #Memory


```
